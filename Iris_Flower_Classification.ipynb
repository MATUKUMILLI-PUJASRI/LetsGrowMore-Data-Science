```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Species_Flower'] = le.fit_transform = (df['Species_Flower'])
df.head(7)
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=['Species_Flower'])
Y = df['Species_Flower']
x_train , x_test , y_train , y_test = train_test_split(X , Y , test_size = 0.7)
```

Logistic Regression

```python
lg= LogisticRegression(max_iter=20)
```

```python
lg.fit(x_train,y_train)
```

```python
y_pred=lg.predict(x_test)
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

```python
print(confusion_matrix(y_pred,y_test))
```

SVC Model

```python
IF_sv= SVC()
IF_sv.fit(x_train,y_train)
L_svc=LinearSVC()
```

```python
L_svc.fit(x_train,y_train)
```

```python
y_pred=L_svc.predict(x_test)
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
print('Linear SVC')
print(classification_report(y_pred,y_test))
```

```python
print(confusion_matrix(y_pred,y_test))
```

Decision Tree Classifier

```python
DTC = DecisionTreeClassifier()
```

```python
DTC=DTC.fit(x_train,y_train)
```

```python
y_pred=DTC.predict(x_test)
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
print('Decision Tree Classifier')
print(classification_report(y_pred,y_test))
```

```python
print(confusion_matrix(y_pred,y_test))
```

Random Forest Classifier

```python
If=RandomForestClassifier(n_estimators=200)
```

```python
If.fit(x_train,y_train)
```

```python
y_pred=If.predict(x_test)
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
print('Decision Tree Classifier')
print(classification_report(y_pred,y_test))
```

Naive Baye's

```python
NB= GaussianNB()
```

```python
NB.fit(x_train,y_train)
```

```python
y_pred=NB.predict(x_test)
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
print('NB')
print(classification_report(y_pred,y_test))
```
